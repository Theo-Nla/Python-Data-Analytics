# Data Cleaning

```python
#Drop last six columns as they will not be used for this analysis

df.drop([
'Gross tertiary education enrollment (%)', 'Population','Unemployment rate','Urban_population','Latitude','Longitude','subscribers_for_last_30_days'], axis = 1, inplace = True)
```

```python
#Replace missing values with not available

df[['category', 'Abbreviation', 'channel_type', 'Country']] = df[['category', 'Abbreviation', 'channel_type', 'Country']].fillna('Not available')
```

```python
df.head(20)
```

```python
#Replace missing value with the mode
# 'country_rank', 'channel_type_rank' , 'video_views_for_the_last_30_days'


df['video_views_for_the_last_30_days'] = df['video_views_for_the_last_30_days'].fillna(df['video_views_for_the_last_30_days'].mode()[0])
df['country_rank'] = df['country_rank'].fillna(df['country_rank'].mode()[0])
df['channel_type_rank'] = df['channel_type_rank'].fillna(df['channel_type_rank'].mode()[0])
```

```python
df.isnull().sum()
```

```python
#Dropping rermaining missing values

df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
# Changing the created_date column to date time

df.created_date = pd.to_datetime(df.created_date)
```

```python
def result(Country):
    if Country in Africa: return 'Africa'
    elif Country in Europe: return 'Europe'
    elif Country in Oceania: return 'Oceania'
    elif Country in South_America: return 'South_America'
    elif Country in North_America: return 'North_America'
    elif Country in Asia: return 'Asia'

#Adding the function into the dataframe

df_clean['Continent'] = df_clean.Country.apply(lambda x:result(x))
```
